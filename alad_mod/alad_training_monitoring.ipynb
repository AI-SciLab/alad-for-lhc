```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
metrics = np.load('/home/oliverkn/pro/results/4_4/alad/21/metrics.npy', allow_pickle=True).item()

fig, ax = plt.subplots(figsize=(10,8))
ax.plot(metrics['epoch'],metrics['auroc'], 'b',label='auroc')
ax.set_ylabel('auroc', color='b')
ax.set_title('AUROC vs recon_loss_diff')
ax.set_xlabel('Epoch')

ax2 = ax.twinx()
ax2.plot(metrics['epoch'], metrics['recon_loss_diff'], 'r')
ax2.set_ylabel('recon_loss_diff', color='r')
fig.show()

fig, ax = plt.subplots(figsize=(10,8))
ax.plot(metrics['epoch'], metrics['recon_loss_sm'], label='recon_loss_sm')
ax.plot(metrics['epoch'], metrics['recon_loss_bsm'], label='recon_loss_bsm')
ax.plot(metrics['epoch'], metrics['recon_loss_train'], label='recon_loss_train')

error = 1

ax.fill_between(metrics['epoch'], np.array(metrics['recon_loss_sm'])-error,np.array(metrics['recon_loss_sm'])+error,
    alpha=0.5, edgecolor='#CC4F1B', facecolor='#FF9848',
    linewidth=1, linestyle='-', antialiased=True)

ax.set_title('Reconstruction loss')
ax.set_xlabel('Epoch')
ax.legend()
fig.show()

#TODO: y-scale: set precentile (cut out outliers)
```
