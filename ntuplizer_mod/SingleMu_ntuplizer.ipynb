```python
part = {'Muons':{}, 'PFCands':{}}
branch = {'Muons':'recoMuons_muons__RECO',
#           'Electrons':'recoGsfElectrons_gsfElectrons__RECO',
          'PFCands':'recoPFCandidates_particleFlow__RECO'
         }

for c in part.keys():
    if c == 'PFCands':
        feat_list = ['pt', 'eta', 'phi', 'pdgId']
    if c == 'Muons':
        feat_list = ['pt', 'eta', 'phi', 'iso']
        
    for fe in feat_list:
        print(c, fe)
        if fe == 'iso':
            aux = 'isolationR03_.sumPt'
        else:
            aux = fe + '_'
        b_name = branch[c] + '.obj.' + aux
        print(b_name)
        part[c][fe] = f['Events'].array(b_name)
```

We are interested in events with at least one muon with $p_T > 23$ and $ISO_{PF} < 0.45$

```python
N_triggerd = 0
for mu_pt, mu_iso in zip(part['Muons']['pt'], part['Muons']['iso']):
    mu_pt = np.array(mu_pt) 
    mu_iso = np.array(mu_iso)/mu_pt
    
    triggered = np.sum(np.logical_and(mu_pt > 23, mu_iso < 0.45))
    N_triggerd += triggered
    
eff = float(N_triggerd)/part['Muons']['pt'].shape[0]
print(eff)
```
