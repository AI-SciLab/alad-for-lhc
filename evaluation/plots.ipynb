```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
metrics = np.load('/home/oliverkn/pro/results/1_1/mlp_binary/1/valid/metrics.npy', allow_pickle=True).item()
```

```python
fig, ax = plt.subplots(figsize=(10,8))
ax.plot(metrics['epoch'],metrics['auc'], label='auroc')
ax.set_title('Training metrics')
ax.set_xlabel('Epoch')
ax.legend()
```

```python
epoch = 5

fig, ax = plt.subplots(figsize=(10,8))
ax.loglog(metrics['roc'][epoch][0],metrics['roc'][epoch][1])
ax.set_title('ROC (epoch %s)'%epoch)
ax.set_xlabel('FPR')
ax.set_ylabel('TPR')

print('AUROC=%s'%metrics['auc'][epoch])
```
